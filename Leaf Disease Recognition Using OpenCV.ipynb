```python
import keras
import tensorflow as tf
import tensorflow.keras.preprocessing.image
from keras.models import load_model
from keras.models import model_from_json
import numpy as np
import cv2
json_file = open('model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
model = model_from_json(loaded_model_json)
model.load_weights('model.h5')
print("Model Loaded Successfully")
cam = cv2.VideoCapture(0)

while (True):
    frame = cam.read()[1]
    img = cv2.resize(frame, (128, 128))

    test_image = tensorflow.keras.preprocessing.image.img_to_array(img)
    test_image = np.expand_dims(test_image, axis=0)
    result = model.predict(test_image)
    arr = np.array(result[0])
    #print(arr)
    max = np.amax(arr)
    max_prob = arr.argmax(axis=0)
    #print(max_prob)
    max_prob = max_prob + 1
    classes = ['healthy', 'algal leaf', 'Anthracnose','brown blight','bird eye spot','gray light','red leaf spot','white spot']
    result = classes[max_prob - 1]
    print(max_prob)
    print(result)

    cv2.putText(frame, result, (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 2)
    cv2.imshow("Leaf", frame)
    key = cv2.waitKey(10)
    if key == 27:
        break
cam.release()
cv2.destroyAllWindows()
```
